## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
num_unique_items = len(purchase_data["Item ID"].unique())
avg_price = purchase_data["Price"].mean()
num_purchases = purchase_data["SN"].count()
tot_revenue = purchase_data["Price"].sum()

purchase_analysis = pd.DataFrame({ "Number of Unique Items" : [num_unique_items],
                         "Average Price" : [avg_price],
                         "Number of Purchases" : [num_purchases],
                         "Total Revenue" : [tot_revenue]
                                 })

purchase_analysis["Average Price"] = purchase_analysis["Average Price"].map("${:.2f}".format)
purchase_analysis["Total Revenue"] = purchase_analysis["Total Revenue"].map("${:.2f}".format)

purchase_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_demo_df = purchase_data

male_count = gender_demo_df[gender_demo_df["Gender"] == "Male"]["SN"].nunique()
female_count = gender_demo_df[gender_demo_df["Gender"] == "Female"]["SN"].nunique()
others_count = gender_demo_df[gender_demo_df["Gender"] == "Other / Non-Disclosed"]["SN"].nunique()
male_percent = ((male_count/total_players)*100)
female_percent = ((female_count/total_players)*100)
others_percent = ((others_count/total_players)*100)

gender_demo_results_df = pd.DataFrame({ "Gender" : ["Male","Female","Other / Non-Disclosed"],
                         "Total Count" : [male_count, female_count, others_count],
                         "Percentage of Players" : [male_percent, female_percent, others_percent]
                                 })

gender_demo_results_df["Percentage of Players"] = gender_demo_results_df["Percentage of Players"].map("{:.2f}".format)

#gender_demo_results_df.reset_index()
gender_demo_results_df.set_index('Gender',inplace=True)
gender_demo_results_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
pur_analysis_df = purchase_data

male_purc_cnt   = pur_analysis_df [pur_analysis_df["Gender"] == "Male"]["Price"].count()
female_purc_cnt = pur_analysis_df [pur_analysis_df["Gender"] == "Female"]["Price"].count()
others_purc_cnt = pur_analysis_df [pur_analysis_df["Gender"] == "Other / Non-Disclosed"]["Price"].count()

male_avg_purc   = pur_analysis_df [pur_analysis_df["Gender"] == "Male"]["Price"].mean()
female_avg_purc = pur_analysis_df [pur_analysis_df["Gender"] == "Female"]["Price"].mean()
others_avg_purc = pur_analysis_df [pur_analysis_df["Gender"] == "Other / Non-Disclosed"]["Price"].mean()

male_tot_purc   = pur_analysis_df [pur_analysis_df["Gender"] == "Male"]["Price"].sum()
female_tot_purc = pur_analysis_df [pur_analysis_df["Gender"] == "Female"]["Price"].sum()
others_tot_purc = pur_analysis_df [pur_analysis_df["Gender"] == "Other / Non-Disclosed"]["Price"].sum()

avg_per_male   = male_tot_purc/male_purc_cnt
avg_per_female = female_tot_purc/female_purc_cnt
avg_per_others = others_tot_purc/others_purc_cnt

pur_analysis_results_df = pd.DataFrame({ "Gender" : ["Male","Female","Other / Non-Disclosed"],
                         "Purchase Count" : [male_purc_cnt, female_purc_cnt, others_purc_cnt],
                         "Average Purchase Price" : [male_avg_purc, female_avg_purc, others_avg_purc],
                         "Total Purchase Value" : [male_tot_purc, female_tot_purc, others_tot_purc],
                         "Avg Total Purchase per Person" : [avg_per_male, avg_per_female, avg_per_others]
                                 })

pur_analysis_results_df["Average Purchase Price"] = pur_analysis_results_df["Average Purchase Price"].map("${:.2f}".format)
pur_analysis_results_df["Total Purchase Value"] = pur_analysis_results_df["Total Purchase Value"].map("${:.2f}".format)
pur_analysis_results_df["Avg Total Purchase per Person"] = pur_analysis_results_df["Avg Total Purchase per Person"].map("${:.2f}".format)

#pur_analysis_results_df.reset_index(drop=True)
#pur_analysis_results_df.set_index('Gender')
#pur_analysis_results_df.groupby(['Gender'], as_index = True)
pur_analysis_results_df.set_index('Gender',inplace=True)
pur_analysis_results_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
topspent_df = purchase_data

tot_pur_val     = pd.DataFrame(topspent_df.groupby('SN')['Price'].sum())
purc_cnt        = pd.DataFrame(topspent_df.groupby('SN')['Price'].count())
ave_purch_price = pd.DataFrame(topspent_df.groupby('SN')['Price'].mean())

top_spent = pd.merge(tot_pur_val, purc_cnt, left_index = True, right_index = True).merge(ave_purch_price, left_index=True, right_index=True)
top_spent.rename(columns = {'Price_x': 'Total Purchase Value', 'Price_y':'Purchase Count', 'Price':'Average Purchase Price'}, inplace = True)
top_spent.sort_values('Total Purchase Value', ascending = False, inplace=True)
top_spent.head(5)

#top_spent["Average Purchase Price"] = top_spent["Average Purchase Price"].map("${:.2f}".format)
#top_spent["Total Purchase Value"] = top_spent["Total Purchase Value"].map("${:.2f}".format)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
topitems_df = purchase_data

popu_items = pd.DataFrame(topitems_df.groupby('Item ID')['Item ID'].count())
popu_items.sort_values('Item ID', ascending = False, inplace = True)
popu_tot = pd.DataFrame(topitems_df.groupby('Item ID')['Price'].sum())
popu_items_merged = pd.merge(popu_items, popu_tot, left_index = True, right_index = True)
drop_items = topitems_df.drop_duplicates(['Item ID'], keep = 'last')
top_item = pd.merge(popu_items_merged, drop_items, left_index = True, right_on = 'Item ID')
top_item = top_item[['Item ID', 'Item Name', 'Item ID_x', 'Price_y', 'Price_x']]
top_item.rename(columns =  {'Item ID_x': 'Purchase Count', 'Price_y': 'Item Price', 'Price_x': 'Total Purchase Value'}, inplace=True)
top_item.head(5)

#disp_items = top_item.set_index('Item ID',inplace=True)
#disp_items.head(5)
```
